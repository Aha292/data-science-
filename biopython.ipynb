```python
import Bio
print(Bio.__version__)
```

```python
from Bio.Seq import Seq

dna = Seq("ATGCGTACGTTAG")
print("DNA:", dna)
print("Complement:", dna.complement())
print("Reverse Complement:", dna.reverse_complement())
print("Transcription (DNA → RNA):", dna.transcribe())
print("Translation (RNA → Protein):", dna.transcribe().translate())
```

```python
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

record = SeqRecord(
    Seq("ATGCGTACGTTAG"),
    id="Seq1",
    description="Sample DNA sequence"
)

print(record.id, record.description, record.seq)
```
